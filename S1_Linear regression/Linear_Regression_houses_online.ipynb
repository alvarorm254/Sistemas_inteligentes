## 1. Reading Data using Panda

```python
# conventional way to import pandas
import pandas as pd
import numpy as np
```

```python
# read CSV file directly from a URL and save the results
# use .read_csv method and simply pass in the name of the files (local and through a url)
# to find out more about this method, click on the url and press shift + tab (twice)
# we set the index_col=0
#data = pd.read_csv('data/train.csv', keep_default_na=False, index_col=0)
data = pd.read_csv('data/train.csv', index_col=0)
Y2=data['SalePrice']
data=data.drop(columns=['SalePrice'])
print(data.shape)

tdata = pd.read_csv('data/test.csv', index_col=0)
print('test: ',tdata.shape)
#tdata.head()
frames = [data, tdata]
adata = pd.concat(frames)
#print(adata.shape)
# display the first 5 rows
#adata.head()
adata.head()
```

## 2. Completando información en la data

```python
# Se analiza cada variable y se remplazan los datos nulos
# por la moda para strings y por la media para otros tipos de datos.
for var in adata:
        if adata[var].dtype!='O':
            adata[var].fillna((adata[var].mean()), inplace=True)
        else:
            adata[var].fillna((adata[var].mode()[0]), inplace=True)
adata_m=adata
adata_m.head()
```

## 4. Spliting data

### 4.1 Input-output

```python
Y=(Y2-Y2.min())/(Y2.max()-Y2.min())
print(Y2.max()-Y2.min())
print(Y2.min())
Y=Y.iloc[:].values;
Y=np.reshape(Y, (Y.shape[0], 1))
X=data2.iloc[:,:].values;
```

### 4.2 Training-Testing

```python
ordering= np.arange(X.shape[0])
np.random.shuffle(ordering)
```

```python
train=ordering[0:int(X.shape[0]*0.8)]
test=ordering[int(X.shape[0]*0.8):X.shape[0]]
#print("train",train.shape)
#print("test",test.shape)
X_train=X[train,:]
X_test=X[test,:]
Y_train=Y[train]
Y_test=Y[test]
```

## 5. Training

```python
############################################
# PENDIENTE
# normal equation : (X'X)^-1.X'Y
############################################
```

```python
############################################
# BATCH
############################################
alfa=0.00001
reg=0.00001
epochs=100000
W=np.random.rand(1,X.shape[1])/X.shape[1]
id=0
for epoch in range(epochs):
    err_ep=0
    for indd in range(X_train.shape[0]):
        err=Y_train[indd]-np.matmul(X_train[indd],np.transpose(W))
        W+=alfa*(err*X_train[indd]+reg*W)
    if epoch%100==0:
        y_pred=np.matmul(X_test,np.transpose(W))*(Y2.max()-Y2.min())+Y2.min()
        y_test=Y_test*(Y2.max()-Y2.min())+Y2.min()
        err_ep=np.sqrt(np.mean((np.log(1+y_pred) - np.log(1+y_test))**2));
        print('Epoca: ',epoch)
        print('err_t: ',err_ep)
print(W)
        
            
            
 
    #err=np.transpose(Y_train)-np.matmul(W,np.transpose(X_train))
 #   dw=alfa*np.matmul(err,X_train)/X.shape[0]
  #  W+=dw
   # #print('Y_train:', err)
    #
    #err_ep=0
    #print('err_ep: ',err_ep) 
   # if epoch%1000==0:
    #    y_pred=np.matmul(X_test,np.transpose(W))*(Y2.max()-Y2.min())+Y2.min()
     #   y_test=Y_test*(Y2.max()-Y2.min())+Y2.min()
      #  err_ep=np.sqrt(np.mean((np.log(1+y_pred) - np.log(1+y_test))**2));
       # print('Epoca: ',epoch)
        #print('err_t: ',err_ep)
#print(W)
```
